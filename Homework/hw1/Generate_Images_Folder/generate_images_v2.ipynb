```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import random

random.seed(3500)
```

```python
def generate_random_image(width, height):
    random_colors = np.random.randint(0, 255, (height, width, 3), dtype=np.uint8)
    return random_colors
```

```python
def display_image(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def download_image(image):
    imageio.imwrite(f'{image.shape[0]}x{image.shape[1]}.png', image)
```

```python
# Example
n = 200
image = generate_random_image(n, n)
display_image(image)
```

```python
n = [10, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

for size in n:
    image = generate_random_image(size, size)
    download_image(image)
```
